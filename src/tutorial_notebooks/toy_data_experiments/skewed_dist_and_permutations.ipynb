```python
import numpy as np
from scipy.stats import skewnorm
import matplotlib.pyplot as plt

# define the parameters
num_values = 10000 # number of random values
skewness = -5      # negative skewness

# create a skew distribution using scipy.stats
random_values = skewnorm.rvs(a = skewness, size=num_values)  

# add some location and scale
random_values = random_values * 10 + 100 # 10 is scale (std dev), 100 is location (mean)
random_values.sort()
```

```python
plt.plot(random_values)
```
